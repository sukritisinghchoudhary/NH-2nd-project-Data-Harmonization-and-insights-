# Box Plot

```python
# Visualize distributions and detect outliers using boxplots 

plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(4, 4, i)
    sns.boxplot(final_data[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
# Now we are detecting statistical outlier detection using IQR

outlier_summary = {}
for col in numeric_columns:
    Q1 = final_data[col].quantile(0.25)
    Q3 = final_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = final_data[(final_data[col] < lower_bound) | (final_data[col] > upper_bound)]
    outlier_summary[col] = len(outliers)

outlier_summary = {k: v for k, v in outlier_summary.items() if v > 0}
outlier_summary
```

NOTE : so here we deduct some outliers and now we are going to handle these outliers using capping method based on IQR.

1 - weathersit: 1 outlier,
2 - temp: 66 outliers,
3 - atemp: 41 outliers,
4 - windspeed: 4 outliers,
5 - casual: 66 outliers,
6 - registered: 45 outliers,
7 - cnt: 30 outliers,

```python
# Handle Outliers using IQR capping

final_data_capped = final_data.copy()

for col in outlier_summary.keys():
    Q1 = final_data[col].quantile(0.25)
    Q3 = final_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
```

```python
# Analyze Skewness of Numerical Columns

numeric_columns = final_data_capped.select_dtypes(include=['float64', 'int64']).columns
skewness = final_data_capped[numeric_columns].apply(lambda x: skew(x.dropna()))

print("Skewness of Numerical Columns:\n", skewness)
```
